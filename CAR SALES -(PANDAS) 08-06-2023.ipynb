## Plotting with Matplolib & Seaborn

```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
data= pd.read_csv('./Car_sales (2).csv')
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().mean()*100
```

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

## Univariate Analysis 

### Histogram

```python
sns.histplot(data=data,x='Sales_in_thousands')
plt.show()
```

```python
sns.boxplot(data=data,x='Sales_in_thousands')
plt.show()
```

```python
sns.boxplot(data=data,y='Sales_in_thousands')
plt.show()
```

```python
sns.kdeplot(data=data.Sales_in_thousands)
plt.show()
```

```python
sns.histplot( data=data)
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.xticks(rotation=75)
sns.boxplot(data=data)
plt.show()
```

```python
sns.kdeplot(data=data)
plt.show()
```

## For loop to plot subplots

```python
for i in data.columns:
    if data[i].dtypes !='object':
        sns.boxplot(y=data[i])
        plt.show()
```

```python
for i in data.columns:
    if data[i].dtypes !='object':
        sns.histplot(x=data[i])
        plt.show()
```

```python
for i in data.columns:
    if data[i].dtypes != 'object':
        sns.kdeplot(data=data[i])
        plt.show()
```

## Subplots in matplotlib

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.boxplot(data=data,y='Sales_in_thousands')

plt.subplot(2,2,2)
sns.boxplot(data=data,y='Price_in_thousands')

plt.subplot(2,2,3)
sns.boxplot(data=data,y='Engine_size')

plt.subplot(2,2,4)
sns.boxplot(data=data,y='Horsepower')

plt.show()
```

```python
features=[]
```

```python
for i in data.columns:
    if data[i].dtypes != 'object':
        features.append(i)
```

```python
features
```

```python
list(enumerate(features))
```

```python
plt.figure(figsize=(13,10))
for i in enumerate(features):
    plt.subplot(3,4,i[0]+1)
    sns.boxplot(data=data,y=i[1],color='red')
    plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('Year Resale Values')
plt.ylabel('Price')
plt.title('GRAPH')
sns.lineplot(data=data,x='__year_resale_value',y='Price_in_thousands')
plt.show()
```

## Count plot 

```python
plt.figure(figsize=(16,5))
plt.xticks(rotation=90)
sns.countplot(data=data,x='Manufacturer')
plt.show()
```

## Area chart 

```python
plt.figure(figsize=(18,5))
plt.xticks(rotation=90)
plt.stackplot(data.Manufacturer,data.Sales_in_thousands)
plt.show()
```

```python
plt.figure(figsize=(18,5))
plt.xticks(rotation=90)
plt.stackplot(data.Manufacturer,data.Sales_in_thousands,data.Price_in_thousands)
plt.show()
```

## Horizontal bar chart

```python
plt.figure(figsize=(6,12))
plt.xticks(rotation=90)
sns.barplot(data=data,x='Sales_in_thousands',y='Manufacturer')
plt.show()
```

## Vertical Bar plot 

```python
plt.figure(figsize=(12,4))
plt.xticks(rotation=90)
sns.barplot(data=data,x='Manufacturer',y='Horsepower')
plt.show()
```

## Scatter plot

```python
sns.scatterplot(data=data,x='Horsepower',y='Sales_in_thousands')
plt.show()
```

```python
sns.scatterplot(data=data,x='Horsepower',y='Sales_in_thousands',hue='Vehicle_type')
plt.show()
```

```python
data.Vehicle_type.value_counts().index
```

```python
palatte_colour= sns.color_palette('dark')
```

```python
plt.pie(data=data,x=data.Vehicle_type.value_counts(),labels=["Passenger", "Car"],explode=[0.1,0],colors=palatte_colour,autopct='%1.2f%%')
plt.legend(bbox_to_anchor = (1.2, 1))
plt.show()
```
