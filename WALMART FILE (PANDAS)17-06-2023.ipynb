## Lets creat a new column called Heat

```python
data['celsius']=[(Temp-32)*(5/9)for Temp in data.Temperature]
```

```python
data['Heat']= pd.cut(x=data.celsius,bins=[-20,10,35],labels=['Cold','Warm'])
```

```python
data
```

```python
sns.barplot(x='Heat',y='WeeklySales',hue='HolidayFlag',data=data)
plt.title("WeeklySales vs Heat")
plt.show()
```
